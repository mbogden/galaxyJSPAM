```python
labels = []
with open('labels.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter='\n')
    for row in readCSV:
        labels.append(row[0])
#labels = []
#for row in readcsv:
    #print(row)
    #labels = labels.append(row)
```

```python
Y = keras.utils.to_categorical(labels)
Y.shape
```

```python
data_train = np.concatenate((good_train, bad_train))
data_test = np.concatenate((good_test, bad_test))
data_train_labels = Y[ : 150]
data_test_labels = Y[150:]
print(data_train.shape)
print(data_test.shape)
print(data_train_labels.shape)
print(data_test_labels.shape)
```

```python
# Mean-center all image data
#X = (X - X.mean()) / X.std()
data_train = (data_train - data_train.mean()) / data_train.std()
data_test = (data_test - data_test.mean()) / data_test.std()
```

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(64, kernel_size=(3,3),
                              activation='relu',
                              input_shape=[data_train.shape[1],
                                           data_train.shape[2],
                                           data_train.shape[3]]))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128,128,1)))
model.add(keras.layers.ELU())
model.add(keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128,128,1)))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(keras.layers.Flatten())




model.add(keras.layers.Dense(2, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(lr = 0.0001),
              metrics=['accuracy'])
model.summary()
```

```python
batch_size = 5
epochs = 5
history = model.fit(data_train, data_train_labels,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_split = 0.2)
```

```python
score = model.evaluate(data_test, data_test_labels, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
